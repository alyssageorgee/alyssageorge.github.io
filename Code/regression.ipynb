```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
a1 = pd.read_excel ('All dataset.xlsx')
```

```python
a1['Anxiety 2019, where 10 = very anxious'].replace(np.nan, a1['Anxiety 2019, where 10 = very anxious'].mean(), inplace=True)
```

```python
linregress(a1['Median Household Income 2019 (annual)'], a1['Anxiety 2019, where 10 = very anxious'])
```

```python
sns.set_theme()
ax=sns.lmplot(x="Median Household Income 2019 (annual)", y="Anxiety 2019, where 10 = very anxious", data=a1).set(title='Anxiety survey score against Household income')
plt.legend(loc='lower right',labels=['y=2.81+1.06*10^-5 x'])
plt.savefig('anxiety vs income.png')
```
