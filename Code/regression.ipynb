```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
a1 = pd.read_excel ('All dataset.xlsx')
```

```python
a1['Suicide Rates (2019) (in %)'].replace(np.nan, a1['Suicide Rates (2019) (in %)'].mean(), inplace=True)
```

```python
linregress(a1['Median Household Income 2019 (annual)'], a1['Suicide Rates (2019) (in %)'])
```

```python
sns.set_theme()
ax=sns.lmplot(x="Median Household Income 2019 (annual)", y="Suicide Rates (2019) (in %)", data=a1).set(title='Suicide rate against Household income')
plt.legend(loc='lower right',labels=['y=-2.72e-05+2.33e-07 x'])
plt.savefig('suicide vs income.png')
```
