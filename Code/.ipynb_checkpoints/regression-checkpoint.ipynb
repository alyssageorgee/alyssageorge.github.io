```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
a1 = pd.read_excel ('All dataset.xlsx')
```

```python
a1['Anxiety 2019, where 10 = very anxious'].replace(np.nan, a1['Anxiety 2019, where 10 = very anxious'].mean(), inplace=True)
```

```python
linregress(a1['Mean Household Income 2019 (annual)'], a1['Anxiety 2019, where 10 = very anxious'])
```

```python
sns.set_theme()
ax=sns.lmplot(x="Mean Household Income 2019 (annual)", y="Anxiety 2019, where 10 = very anxious", data=a1).set(title='Mean Household income vs Anxiety survey score')
plt.legend(loc='lower right', labels=['y=3.03+2.12*10^-6 x'])
plt.savefig('anxiety vs income.png')
```
